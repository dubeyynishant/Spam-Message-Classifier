```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.linear_model import LogisticRegression,SGDClassifier
```

```python
df=pd.read_csv(r"D:\Ducat Data Science Professionals\Machine Learning (ML)\ML DataSets\spam_ham.txt",sep='\t')
df
```

```python
df.Target.value_counts()
```

```python
corpus=df.Msg
y=df.Target
```

```python
cv=CountVectorizer(lowercase=True,stop_words='english')
X=cv.fit_transform(corpus)
```

```python
from sklearn.model_selection import cross_validate
```

```python
model=LogisticRegression(class_weight={'ham':1/4825,'spam':1/747})
d=cross_validate(model,X,y,cv=5,return_train_score=True)
print(d['train_score'].mean())
print(d['test_score'].mean())
```

```python
cv=CountVectorizer(lowercase=True,stop_words='english',ngram_range=(1,1),max_df=1.0,min_df=1)
X=cv.fit_transform(corpus)
model=LogisticRegression(class_weight={'ham':1/4825,'spam':1/747})
d=cross_validate(model,X,y,cv=5,return_train_score=True)
print(d['train_score'].mean())
print(d['test_score'].mean())
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
model=KNeighborsClassifier()
d=cross_validate(model,X,y,cv=5,return_train_score=True)
print(d['train_score'].mean())
print(d['test_score'].mean())
```

```python
model=DecisionTreeClassifier()
d=cross_validate(model,X,y,cv=5,return_train_score=True)
print(d['train_score'].mean())
print(d['test_score'].mean())
```

```python
model=RandomForestClassifier()
d=cross_validate(model,X,y,cv=5,return_train_score=True)
print(d['train_score'].mean())
print(d['test_score'].mean())
```

```python
cv=CountVectorizer(lowercase=True,stop_words='english',ngram_range=(1,3))
X=cv.fit_transform(corpus)
model=SGDClassifier(class_weight={'ham':1/4825,'spam':1/747})
d=cross_validate(model,X,y,cv=5,return_train_score=True)
print(d['train_score'].mean())
print(d['test_score'].mean())
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe=Pipeline(
    [('cv',CountVectorizer(lowercase=True,stop_words='english')),
     ('model',model)])
pipe.fit(corpus,y)
```

```python
pipe.predict(['you got selected as a lucky winner','congratulations you won free world tour with family'])
```
